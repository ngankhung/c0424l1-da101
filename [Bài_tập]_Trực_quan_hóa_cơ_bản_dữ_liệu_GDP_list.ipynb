```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/GDPlist.csv', encoding='latin-1')
```

```python
df
```

```python
# So sánh GDP các nước ở South America.

south_america_gdp = df[df['Continent'] == 'South America']
```

```python
south_america_gdp
```

```python
plt.figure(figsize=(10, 6))
plt.bar(south_america_gdp['Country'], south_america_gdp['GDP (millions of US$)'], color='skyblue')
plt.xlabel('GDP (millions of US$)')
plt.title('GDP of South American Countries')
plt.show()
```

```python
# Đánh giá tỉ lệ đóng góp GDP của Việt Nam trên tổng số GDP của 5 nước Đông Nam Á là Vietnam,  Indonesia, Cambodia, Thailand và Malaysia.

asean_countries = ['Vietnam', 'Indonesia', 'Cambodia', 'Thailand', 'Malaysia']
df_asean = df[df['Country'].isin(asean_countries)]

total_gdp_asean = df_asean['GDP (millions of US$)'].sum()

gdp_vietnam = df_asean[df_asean['Country'] == 'Vietnam']['GDP (millions of US$)']
gdp_vietnam_ratio = (gdp_vietnam / total_gdp_asean) * 100
```

```python
plt.figure(figsize=(10, 7))
plt.bar(df_asean['Country'], df_asean['GDP (millions of US$)'])
plt.xlabel('Country')
plt.ylabel('GDP (millions of US$)')
plt.title('GDP của các nước Đông Nam Á năm 2019')
plt.show()
```
