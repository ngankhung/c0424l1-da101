```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/bài tập colab/OnlineRetail.csv', encoding='latin-1')
df.head()
```

```python
# Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
print('Số dòng và số cột:', df.shape)
df.info()
```

```python
# Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv
df = df.loc[:,['Description', 'Quantity']]
df.to_csv('OnlineRetail.csv')
data = pd.read_csv('OnlineRetail.csv')
data.head()
```

```python
# Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx
df = df.iloc[:1000]
df.to_excel('OnlineRetail.xlsx')
data = pd.read_excel('OnlineRetail.xlsx')
data.head()
```

```python
# Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5
df = df.loc[df['Quantity'] >= 10]
df.to_hdf('OnlineRetail.h5', key = 'df')
data = pd.read_hdf('OnlineRetail.h5', key = 'df')
data.head()
```

```python
# Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json
df = df.loc[1000:2000,['Quantity','InvoiceNo','UnitPrice']]
df.to_json('OnlineRetail.json', orient = 'columns')
data = pd.read_json('OnlineRetail.json')
data.head()
```

```python
# Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html
df = df.loc[df['InvoiceNo'] == '536365']
df.to_html('OnlineRetail.html')
data = pd.read_html('OnlineRetail.html')
data
```
