#Có nên xóa hết các dòng chứa dữ liệu khuyết thiếu, giải thích vì sao?

Việc xóa các dòng chứa dữ liệu khuyết thiếu có thể làm mất nhiều thông tin quan trọng, đặc biệt khi dữ liệu thiếu nhiều. Điều này có thể làm giảm kích thước dữ liệu và ảnh hưởng đến kết quả phân tích. Thay vì xóa, ta có thể thay thế các giá trị khuyết thiếu bằng giá trị mặc định

```python
# Thực hiện xử lý giá trị khuyết thiếu: Thay thế giá trị khuyết thiếu của thuộc tính Description bằng giá trị mặc định “Không biết”
df['Description'] = df['Description'].fillna('Không biết', inplace=True)
```

```python
# Thực hiện phát hiện giá trị ngoại lai của thuộc tính Quantity và Thuộc tính UnitPrice

# Sử dụng phương pháp IQR để phát hiện giá trị ngoại lai
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1

quantity_outliers = df[(df['Quantity'] < (Q1 - 1.5 * IQR)) | (df['Quantity'] > (Q3 + 1.5 * IQR))]

Q1 = df['UnitPrice'].quantile(0.25)
Q3 = df['UnitPrice'].quantile(0.75)
IQR = Q3 - Q1

unitprice_outliers = df[(df['UnitPrice'] < (Q1 - 1.5 * IQR)) | (df['UnitPrice'] > (Q3 + 1.5 * IQR))]
```

```python
# Dữ liệu ngoại lai
outliers = df[(df['Quantity'].isin(quantity_outliers['Quantity'])) | (df['UnitPrice'].isin(unitprice_outliers['UnitPrice']))]

# Dữ liệu sạch
clean_data = df[~df.index.isin(outliers.index)]
```
