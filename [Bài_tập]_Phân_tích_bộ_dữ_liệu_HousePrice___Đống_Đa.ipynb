```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_data = pd.read_csv('/content/house_price_Dống-Da_Hà-Nội_subdata1.csv')
house_data
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(house_data['area'], house_data['price'], alpha=0.5)
plt.title('Area vs. Price')
plt.xlabel('Area (m²)')
plt.ylabel('Price (VND)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(house_data['bedroom'], house_data['price'], alpha=0.5)
plt.title('Bedrooms vs. Price')
plt.xlabel('Bedrooms')
plt.ylabel('Price (VND)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(house_data['toilet'], house_data['price'], alpha=0.5)
plt.title('Toilets vs. Price')
plt.xlabel('Toilets')
plt.ylabel('Price (VND)')
plt.show()
```

```python
house_data['price_per_m2'] = house_data['price'] / house_data['area']
avg_price_per_m2 = house_data.groupby('property_type')['price_per_m2'].mean()
avg_price_per_m2
plt.figure(figsize=(10, 6))
avg_price_per_m2.plot(kind='bar', color='skyblue')
plt.title('Average House Price per Square Meter by Type of Land')
plt.xlabel('Type of Land')
plt.ylabel('Average Price per m² (VND)')
plt.show()
```

```python
property_type_counts = house_data['property_type'].value_counts()
plt.figure(figsize=(10, 6))
property_type_counts.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'yellowgreen'])
plt.title('Percentage of Listings by Type of Land')
plt.ylabel('')
plt.show()
```

```python
avg_price_per_m2_by_bedroom = house_data.groupby('bedroom')['price_per_m2'].mean()
plt.figure(figsize=(10, 6))
avg_price_per_m2_by_bedroom.plot(kind='bar', color='skyblue')
plt.title('Average House Price per Square Meter by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price per m² (VND)')
plt.show()
```
