```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('/content/drive/MyDrive/bài tập colab/CSV/shopeep_koreantop_clothing_shop_data.csv')
```

```python
# Vẽ biểu đồ tần số số lượng shop gia nhập theo các năm.
plt.figure(figsize=(10, 6))
sns.countplot(x='join_year', data=df)
plt.title('Shop joining by year')
plt.xlabel('Year')
plt.ylabel('Shops')
plt.show()
```

```python
# Vẽ biểu đồ xu hướng thể hiện mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt.
plt.figure(figsize=(10, 6))
sns.lmplot(x='response_rate', y='rating_good', data=df)
plt.title('Response rate vs Rating Good')
plt.xlabel('Response Rate')
plt.ylabel('Rating Good')
plt.show
```

```python
# Vẽ biểu đồ xu hướng thể hiện mối quan hệ giữa thời gian phản hồi (đơn vị giây) với số lượt khách hàng đánh giá xấu.
df['seconds'] = pd.to_timedelta(df['response_time']).dt.total_seconds()
plt.figure(figsize=(10, 6))
sns.lmplot(x='seconds', y='rating_bad', data=df)
plt.title('Response time vs Rating Bad')
plt.xlabel('Response Time (seconds)')
plt.ylabel('Rating Bad')
```

```python
# Vẽ biểu đồ thể hiện phân bố của điểm đánh giá trung bình.
plt.figure(figsize=(10, 6))
sns.histplot(df['rating_star'], bins=10, kde=True)
plt.title('Rating Average Distribution')
plt.xlabel('Rating Average')
plt.ylabel('Count')
plt.show
```

```python
# Vẽ biểu đồ tần số của cửa hàng chính thức và không chính thức.
plt.figure(figsize=(10, 6))
sns.countplot(x='is_official_shop', data=df)
plt.title('Official Store Distribution')
plt.xlabel('Official Store')
plt.ylabel('Count')
plt.show
```

```python
# Vẽ biểu đồ tần số của cửa hàng được xác thực với chưa xác thực.
plt.figure(figsize=(10, 6))
sns.countplot(x='is_shopee_verified', data=df)
plt.title('Verified Store Distribution')
plt.xlabel('Verified Store')
plt.ylabel('Count')
plt.show
```
