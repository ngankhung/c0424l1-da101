```python
import pandas as pd
```

```python
online_df = pd.read_csv('/content/drive/MyDrive/bài tập colab/CSV/OnlineRetail.csv', encoding='latin-1')
```

```python
online_df.head()
```

```python
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
online_df.shape
```

```python
online_df.info()
```

```python
#Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia.
online_df.pivot_table(values='Quantity',
                      index='InvoiceNo',
                      columns='Country',
                      aggfunc='mean')
```

```python
#Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.
online_df.pivot_table(values='Quantity',
                      index='CustomerID',
                      columns='StockCode',
                      aggfunc={'min','max'})
```

```python
#Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.
online_df.pivot_table(values=['Quantity','UnitPrice'],
                      index='StockCode',
                      aggfunc={'Quantity':'sum','UnitPrice':'mean'})
```

```python
#Xây dựng bảng Pivot table cho biết tổng số lượng hàng bán được của mỗi ngày.
online_df.pivot_table(values='Quantity',
                      index='InvoiceDate',
                      aggfunc='sum')
```
