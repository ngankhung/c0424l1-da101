```python
# Các giá trị GDP ban đầu dưới dạng scientific notation
gdp_values = {
    'count': 1.250000e+02,
    'mean': 5.554278e+05,
    'std': 1.649266e+06,
    'min': 1.060000e+04,
    '25%': 2.574200e+04,
    '50%': 9.608900e+04,
    '75%': 3.332380e+05,
    'max': 1.509402e+07
}

# Quy đổi và hiển thị các giá trị
converted_gdp_values = {key: "{:,.1f}".format(value) for key, value in gdp_values.items()}

# In ra kết quả
for key, value in converted_gdp_values.items():
    print(f"{key}: {value}")
```

```python
# mean: 555,427.8: trung bình GDP
# std: 1,649,266.0: độ lệch chuẩn so với trung bình cao cho thấy GDP của một số quốc gia rất cao
# min: 10,600.0: giá trị nhỏ nhất
# max: 15,094,020.0: giá trị lớn nhất
# 25%: 25,742.0: quantile 1 tức là phân vị 1 có GDP <= 25,742.0 $
# 50%: 96,089.0: quantile 2 tức là phân vị 2 có GDP <= 96,089.0 $
# 75%: 333,238.0: quantile 3 tức là phân vị 3 có GDP <= 333,238.0 $
# => GDP ở các quốc gia có sự chênh lệch rất lớn, bộ dữ liệu có outlier vì độ lệch chuẩn của GDP so với trung bình GDP là rất cao
```

```python
# Hãy cho biết châu lục nào xuất hiện nhiều nhất?
continent_df = gdp_df['Continent'].value_counts()
continent_df
```

```python
print('Châu lục xuất hiện nhiều nhất:', continent_df.idxmax())
```

```python
# Với mỗi châu lục hãy tính tổng GDP; trung bình cộng GDP. Hợp nhất 2 bảng này thành một bảng duy nhất gồm 3 thông tin: Tên châu lục; Tổng GDP; TBC GDP.
gdp_sum = gdp_df.groupby('Continent')['GDP (millions of US$)'].sum().reset_index()
```

```python
gdp_sum
```

```python
gdp_mean = gdp_df.groupby('Continent')['GDP (millions of US$)'].mean().reset_index()
```

```python
gdp_mean
```

```python
result = pd.merge(gdp_sum, gdp_mean, on='Continent')
```
