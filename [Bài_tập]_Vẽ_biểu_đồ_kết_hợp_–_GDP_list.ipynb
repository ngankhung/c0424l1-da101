```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/bài tập colab/CSV/GDPlist.csv', encoding="ISO-8859-1")
```

```python
df.head()
```

```python
df['Country'].unique()
```

```python
countries = ['\xa0Vietnam', '\xa0Indonesia', '\xa0Cambodia', '\xa0Thailand', '\xa0Malaysia']
gdp_df = df[df['Country'].isin(countries)]
gdp_df.rename(columns={'GDP (millions of US$)': 'GDP'}, inplace=True)
gdp_df
```

```python
# Vẽ biểu đồ so sánh GDP của 5 nước
plt.figure(figsize=(10, 6))
plt.bar(gdp_df['Country'], gdp_df['GDP'])
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('GDP of 5 Countries')
plt.show()
```

```python
# Tính toán tỉ lệ đóng góp GDP của các nước
total_gdp = gdp_df['GDP'].sum()
gdp_contribution = gdp_df['GDP'] / total_gdp * 100

# Vẽ biểu đồ để đánh giá tỉ lệ đóng góp GDP
plt.figure(figsize=(10, 6))
plt.bar(gdp_df['Country'], gdp_contribution)
plt.xlabel('Country')
plt.ylabel('Contribution to Total GDP (%)')
plt.title('Contribution to Total GDP of 5 Countries')
plt.show()
```

```python
# Vẽ biểu đồ kết hợp
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

ax[0].bar(gdp_df['Country'], gdp_df['GDP'])
ax[0].set_xlabel('Country')
ax[0].set_ylabel('GDP')
ax[0].set_title('GDP of 5 Countries')

ax[1].bar(gdp_df['Country'], gdp_contribution)
ax[1].set_xlabel('Country')
ax[1].set_ylabel('Contribution to Total GDP (%)')
ax[1].set_title('Contribution to Total GDP of 5 Countries')

plt.show()
```
