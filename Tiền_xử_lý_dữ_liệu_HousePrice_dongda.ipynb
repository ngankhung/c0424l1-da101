```python
df = df.dropna(subset=['price'])
```

```python
df.isna().sum()
```

Thực hiện xử lý giá trị khuyết thiếu: Thay thế giá trị khuyết thiếu của land_certificate bằng =”không có thông tin”, house_direction, balcony_direction, toilet, bedroom, Floor  bằng giá trị có tần số xuất hiện lớn nhất của các thuộc tính đó,
Lọc thông tin những bất động sản ở trong ngõ thành bộ dữ liệu nhà ngõ

```python
df['land_certificate'] = df['land_certificate'].fillna("không có thông tin")
df['house_direction'] = df['house_direction'].fillna(df['house_direction'].mode()[0])
df['balcony_direction'] = df['balcony_direction'].fillna(df['balcony_direction'].mode()[0])
df['toilet'] = df['toilet'].fillna(df['toilet'].mode()[0])
df['bedroom'] = df['bedroom'].fillna(df['bedroom'].mode()[0])
df['floor'] = df['floor'].fillna(df['floor'].mode()[0])
```

```python
df.isna().sum()
```

Lọc thông tin những bất động sản ở trong ngõ thành bộ dữ liệu nhà ngõ


```python
trong_ngo = df[df['property_type'] == 'trong ngo']
```

```python
trong_ngo = trong_ngo[trong_ngo['area'] == 0]
```

```python
trong_ngo
```

Tính toán giá/m2  ( đơn vị triệu/m2) với loại hình nhà ngõ


```python
trong_ngo['gia/m2'] = trong_ngo['price']/trong_ngo['area']
```

Phát hiện giá trị ngoại lai của các thuộc tính: diện tích, giá/m2 bằng phương pháp IQR


```python
q3_area = trong_ngo['area'].quantile(0.75)
q1_area = trong_ngo['area'].quantile(0.25)
iqr_area = q3_area - q1_area
outlier_area = trong_ngo[(trong_ngo['area'] > q3_area + 1.5*iqr_area) | (trong_ngo['area'] < q1_area - 1.5*iqr_area)]
```

```python
outlier_area
```

```python
q3 = trong_ngo['gia/m2'].quantile(0.75)
q1 = trong_ngo['gia/m2'].quantile(0.25)
iqr = q3 - q1
outlier = trong_ngo[(trong_ngo['gia/m2'] > q3 + 1.5*iqr) | (trong_ngo['gia/m2'] < q1 - 1.5*iqr)]
```

```python
outlier
```

Thực hiện loại bỏ các dòng dữ liệu ngoại lai

```python
clean_ngo = trong_ngo[~(trong_ngo['area'] > q3_area + 1.5*iqr_area) | (trong_ngo['area'] < q1_area - 1.5*iqr_area)]
```

```python
clean_ngo = clean_ngo[(clean_ngo['gia/m2'] > q3 + 1.5*iqr) | (clean_ngo['gia/m2'] < q1 - 1.5*iqr)]
```
